```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from Analisis_Predictivo import Analisis_Predictivo
from Analisis_Predictivo import Graficar
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis

import os
os.chdir("/home/william/Desktop")
from Analisis_Predictivo import GraficarROC
```

```python
potabilidad = pd.read_csv("potabilidad_V2.csv")
potabilidad = potabilidad.iloc[:,1:]
potabilidad["Potability"] = potabilidad["Potability"].astype('category')
```

```python
decisionTree =  DecisionTreeClassifier()
analisisDecisionTree = Analisis_Predictivo(potabilidad,predecir= "Potability",modelo=decisionTree, train_size= 0.80)
analisisDecisionTree.training_testing_con_reja_corte(corte=[0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90], model=decisionTree)
```

```python
extremeGradiantBoosting = GradientBoostingClassifier()
analisisGradiantBoosting = Analisis_Predictivo(potabilidad,predecir= "Potability",modelo=extremeGradiantBoosting, train_size= 0.80)
analisisGradiantBoosting.training_testing_con_reja_corte(corte=[0.20, 0.30,0.40, 0.50, 0.60, 0.70, 0.80, 0.90], model=extremeGradiantBoosting)
```

```python
bayes = GaussianNB()
knn_model =  KNeighborsClassifier(n_neighbors=60, algorithm="auto")
decisionTree =  DecisionTreeClassifier(criterion="gini")
adaBoosting = AdaBoostClassifier()
extremeGradiantBoosting = GradientBoostingClassifier()
lda = LinearDiscriminantAnalysis()
qda = QuadraticDiscriminantAnalysis()
analisis = Analisis_Predictivo(potabilidad,predecir= "Potability",modelo=lda,
                                       train_size= 0.75)
```

```python
graficarROC = GraficarROC(models=[bayes, knn_model, decisionTree, adaBoosting, extremeGradiantBoosting, lda, qda], label=["bayes", "knn_model", "decisionTree", "adaBoosting", "extremeGradiantBoosting", "lda", "qda"],color=["red","green","purple", "yellow", "blue", "black", "purple"], x=analisis.x, y=analisis.y, predecir="Potability")
graficarROC.graficar()

```
